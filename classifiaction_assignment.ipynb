```python
#Loading the dataset
import os
os.getcwd()
os.chdir('C:\\Users\\user\\Desktop\\awo')
import pandas as pd
df = pd.read_csv("data.csv")
df.head(20)
```

```python
#separation  into x and y values 
x=df.iloc[:,0:-1]
y=df.iloc[:,-1] #roomtype (target values)
```

```python
#training the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
#Decision tree classifier
from sklearn import tree
dt_clf=tree.DecisionTreeClassifier(max_depth=10)
dt_clf.fit(x_train, y_train)
y_pred=dt_clf.predict(x_test)
dt_clf.score(x_test,y_test)
```

```python
#creating the classification report
from sklearn. metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#precision: The proportion of correct identifications among each class.
#recall: The proportion of all the correctly found positive instances.
#f1 score: The harmonic mean between the precision and recall
#support: The number of samples in each classification in each class/metric was tested on.
#accuracy: This is the percentage of labels that the model successfully predicted.
#Macro avg: This is the arithmetic mean of the individual class related to precision,memory,and f1 score.
#Weigthed avg: This is the average considering the proportion for each label in the data set.
```
